#### Visualizar distribuciones de las características numéricas.

#### Se visualiza distribución de las características numericas : 'age', 'balance', 'duration', ´campain', ´pdays', 'previous'

```python
# Crear la figura y los subplots.
fig, axs = plt.subplots(1, 6, figsize=(16, 4))

# Se realiza gráfico de distribución para columna 'age'
axs[0].hist(df['age'], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
axs[0].set_title("Histograma Edad", fontsize=14, fontweight='bold')
axs[0].set_xlabel("age", fontweight='bold')
axs[0].set_ylabel("Frecuencia", fontweight='bold')
axs[0].grid(True, linewidth=0.5)

# Se realiza gráfico de distribución para columna 'balance'
axs[1].hist(df['balance'], bins=10, color='lightgreen', edgecolor='black', alpha=0.7)
axs[1].set_title("Histograma Balance", fontsize=14, fontweight='bold')
axs[1].set_xlabel("balance", fontweight='bold')
axs[1].set_ylabel("Frecuencia", fontweight='bold')
axs[1].grid(True, linewidth=0.5)

# Se realiza gráfico de distribución para columna 'duration'
axs[2].hist(df['duration'], bins=10, color='lightsalmon', edgecolor='black', alpha=0.7)
axs[2].set_title("Histograma Duratión", fontsize=14, fontweight='bold')
axs[2].set_xlabel("duration", fontweight='bold')
axs[2].set_ylabel("Frecuencia", fontweight='bold')
axs[2].grid(True, linewidth=0.5)


# Se realiza gráfico de distribución para columna 'campaign'
axs[3].hist(df['campaign'], bins=10, color='lightcyan', edgecolor='black', alpha=0.7)
axs[3].set_title("Histograma Campaign", fontsize=14, fontweight='bold')
axs[3].set_xlabel("campaign", fontweight='bold')
axs[3].set_ylabel("Frecuencia", fontweight='bold')
axs[3].grid(True, linewidth=0.5)

# Se realiza gráfico de distribución para columna 'pdays'
axs[4].hist(df['pdays'], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
axs[4].set_title("Histograma Pdays", fontsize=14, fontweight='bold')
axs[4].set_xlabel("pdays", fontweight='bold')
axs[4].set_ylabel("Frecuencia", fontweight='bold')
axs[4].grid(True, linewidth=0.5)

# Se realiza gráfico de distribución para columna 'previous'
axs[5].hist(df['previous'], bins=10, color='green', edgecolor='black', alpha=0.7)
axs[5].set_title("Histograma Previous", fontsize=14, fontweight='bold')
axs[5].set_xlabel("previous", fontweight='bold')
axs[5].set_ylabel("Frecuencia", fontweight='bold')
axs[5].grid(True, linewidth=0.5)

# Ajustar el espacio entre gráficos.
plt.tight_layout()
plt.show()
```

* En el gráfico de Edad, se observa que la mayoria de clientes esta en el rango de 20 a 60 años, con valores elevados alrededor de 40 años. Esto indica que el producto o servicio, es más relevante para personas en edad laboral.

* En el gráfico de Balance(Saldo medio anual), se observa que la mayor cantidad de clientes se encuentra con balance igual a 0, la distribución esta en el rango de 0 a 25.000 euros. Una gran parte de clientes no mantienen saldos significativos.

* En el gráfico Duración(Duración del último contacto en segundos), se observa que la mayor cantidad de contactos duraron 0 segundos, en un rango de 0 a 2000 segundos. Podria reflejar intentos fallidos de contacto(Llamadas no contestadas).

* En el gráfico de Campaña(número de contactos realizados durante esta campaña y para este cliente). Se oberva que la mayor frecuencia esta alrededor de 0, en un rango de 0 a 20 contactos. Se deduce que la mayoria de cliente no fue contactado durante esta camapaña.

* En el gráfico de Pdays(Numero de dias que pasaron despues que el cliente fue contactado ppor última vez, de una campaña anterior), Se observa que la frecuencia más alta se da en el valor de 0 dias, en un rango de 0 a 500 dias. Podria indicar contactos recientes.   

* En el gráfico Previous se observa número de contactos en el rango de 0 a 10, con frecuencia elevada alrededor del valor 0, Podria indicar mayoria de clientes no contactados antes de esta campaña.

#### Analizar la correlación entre las características.

#### Para analizar correlaciones entre las características, se muestra un gráfico de correlaciones para ser más explicito.

```python
# Se reviza los datos con la correlaciones de Pearson
# Se filtra solo las columnas numericas

df_numericas =df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = df_numericas.corr(method='pearson')
print(correlation_matrix)
```

```python
# Se Crea el  mapas de calor
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Mapa de Calor de la Correlación de Pearson")
plt.show()
```

#### Correlaciones significativas :
   Relación entre Pday y Previous (0.58). Existe una correlacion positiva moderada, entre dias que pasaron, despues que el cliente fue contactado por última vez de una campaña anterior(Pday) y el número de contactos realizados antes de esta campaña(Previous). Esto indica que clientes contactados en el pasado son más propensos a ser contactados en nuevas campañas.


#### Las mayoria de variables muestran correlaciones insignificantes, Esto demuestra independencia entre las características.
