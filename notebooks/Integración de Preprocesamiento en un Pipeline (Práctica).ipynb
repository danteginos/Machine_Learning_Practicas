#### Objetivo: Integrar diferentes técnicas de preprocesamiento en un pipeline completo.

* Instrucciones

 1. Carga del Dataset:
    Utilizar el dataset Wine Quality de Scikit-learn.
 2. Tareas:

    Manejar valores faltantes.
    Codificar variables categóricas.
    Escalar características numéricas.
    Integrar todas las transformaciones en un pipeline.


```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import numpy as np
```

```python
# Se carga  el dataset Wine Quality
wine = load_wine()
X = pd.DataFrame(wine.data, columns=wine.feature_names)
y = wine.target
```

```python
# Se  añade valores faltantes para la práctica
X.loc[0:10, 'alcohol'] = np.nan
```

```python
# Se define las transformaciones para imputar datos numericos y escalarlos
numeric_features = X.select_dtypes(include=['float64', 'int']).columns
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])
```

```python
# En este dataset no hay categóricas, pero se puede añadir una columna categórica ficticia para la práctica
# Se codifica las variables categoricas con el onehotencoder
X['quality'] = np.where(y > 1, 'high', 'low')
categorical_features = ['quality']
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
# Se  combina las transformaciones con el Columntransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
# Se Integrar en un pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])
```

```python
# Se aplicar preprocesamiento para que las variqables esten listas para el modelamiento
X_transformed = pipeline.fit_transform(X)
```

```python
print("Preprocesamiento completado. Datos transformados listos para modelar.")
```
